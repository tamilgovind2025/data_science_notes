```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a simple line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Matplotlib Plot')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot with customization
plt.plot(x, y, color='green', linestyle='dashed', marker='o', markersize=8, label='Customized Plot')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Matplotlib Plot')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
#plotting sin and cosine wave forms 

import numpy as np 
import matplotlib.pyplot as plt 

%matplotlib inline 

x=np.linspace(0,8,800)
y1=np.sin(x)
y2=np.cos(x)

fig=plt.figure()

plt.plot(x,y1,'-')
plt.plot(x,y2,'--')

```

```python
#drawing with out line 
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,5,8,12,6,8])
ypoints = np.array([3,7,8,9,21,10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
# a figure with one axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
```
